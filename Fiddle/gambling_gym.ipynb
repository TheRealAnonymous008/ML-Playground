```python
%load_ext autoreload
%autoreload 2

import gymnasium as gym
import matplotlib.pyplot as plt
import numpy as np

from rl.policy import Policy
from rl.mc import MonteCarlo
import matplotlib
```

```python
env = gym.make("Blackjack-v1", render_mode = "None")
shape = [i.n for i in env.observation_space]
policy = Policy(shape, env.action_space.n)
mc = MonteCarlo(policy=policy, env = gym.make("Blackjack-v1", render_mode = "None"), sampling_policy=policy)
```

```python
mc.forward(1000000, epsilon=0.05)
```

```python
env = gym.make("Blackjack-v1", render_mode = "None")
episode_count = 100000
state, info = env.reset(seed=42)

ret = 0
for _ in range(episode_count):
    isStop = False

    state, info = env.reset()
    while isStop is False:
        # Sample from a policy
        action = mc.policy.sample(state)

        state, reward, isTerminated, isTruncated, info = env.step(action)

        ret += reward
        if isTerminated or isTruncated:
            isStop = True


    state, info = env.reset()

env.close()

print(ret / episode_count)
```
