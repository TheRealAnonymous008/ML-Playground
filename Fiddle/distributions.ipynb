```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(1e-2, 8, 100)

dfs = np.arange(2, 10, 0.1)
for i, df in enumerate(dfs):
    y = stats.wishart.pdf(x, df=df, scale = 1)

    c = i / len(dfs)
    plt.plot(x, y, c=[c, 0, 1 - c])
plt.title("Wishart Distribution")
plt.legend()
plt.show()
```

```python
def generate_grid(xmin, xmax, x_n, ymin, ymax, y_n):
    xs = np.linspace(xmin, xmax, x_n)
    ys = np.linspace(ymin, ymax, y_n)
    x, y = np.meshgrid(xs, ys)
    xy = np.array([x.flatten(), y.flatten()]).T
    return xy, x, y
```

```python
mesh, xx, yy = generate_grid(-3, 3, 100, -3, 3, 100)
y = stats.multivariate_normal.pdf(
        mesh, 
        mean=[0, 0], 
        cov= [[1, -1.25],
              [-1.25, 2.2]]              
    )

plt.contourf(xx, yy, y.reshape(100, 100), cmap=plt.cm.RdYlBu, alpha=0.7)

plt.title("Multivariate Normal Distribution")
plt.legend()
plt.show()
```
